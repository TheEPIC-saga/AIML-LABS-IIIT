##Multiple dot product##

```python
# matrices with random integers: entries ranging from -5 to 6
a = np.random.randint(-5,6,(500,50))
b = np.random.randint(-5,6,(50,1000))
c = np.random.randint(-5,6,(1000,10))
d = np.random.randint(-5,6,(10,200))
e = np.random.randint(-5,6,(200,200))
# Perform multiple matrix multiplication
z = np.linalg.multi_dot([a,b,c,d,e])
```

```python
%%time
z = np.linalg.multi_dot([a,b,c,d,e])
print(z, '\n')
```

```python
%%time
z = a.dot(b).dot(c).dot(d).dot(e)
print(z, '\n')
```

#Solving System Of Equations

```python
# Coefficient Matrix
a = np.random.randint(1,100,[5,5])
# Dependent variable vector
b = np.array([5,6,3,2,4])
# solution
x = np.linalg.solve(a,b)
print('Coefficient Matrix')
print(a)
print('\nDependent Variable vector')
print(b)
print('\nSolution')
print(x)
```

```python
# Check for correctness
B = a.dot(x)
print(B)
```

#Singular Value Decompositon#

```python
from skimage import data
image = data.coffee()
print(image.dtype, image.min(), image.max(), image.shape)
plt.imshow(image)
plt.show()
```

```python
# normalize image
img = image/255.0
# reorder the axes to have colour channels as the first dimension
img = np.transpose(img, axes=(2,0,1))
```

```python
U,S,V = np.linalg.svd(img)
print(U.shape, S.shape, V.shape)
```

```python
# S matrix should have dimensions suitable for matrix multiplication
Sigma = np.zeros((3,400,600))
for i in range(3):
  np.fill_diagonal(Sigma[i,:,:], S[i,:])
print(Sigma.shape)
```

```python
# Reconstruct the original image without any dimension reduction.
reconst = U @ Sigma @ V
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

```python
k = 20
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

```python
k = 10
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```
